```python
import sensitivity_tests as st
import matplotlib.pyplot as plt
```

```python
def test_noise_gens():
    swp = st.SwapNoise(0.25)
    bs = st.BootstrapResamplingNoise(0.25)
    nrm = st.NormalResamplingNoise(0.25)
    nrm_clp = st.NormalResamplingNoise(0.25, clip_range=True)
    D = st.PerfectWeightedSource(16).init_D()
    plt.title("Original D")
    plt.imshow(D)
    plt.show()
    D_noisy = swp.apply_noise(D)
    plt.title("SwapNoise")
    plt.imshow(D_noisy)
    plt.show()
    D_noisy = bs.apply_noise(D)
    plt.title("BootstrapResamplingNoise")
    plt.imshow(D_noisy)
    plt.show()
    D_noisy = nrm.apply_noise(D)
    plt.title("NormalResamplingNoise")
    plt.imshow(D_noisy)
    plt.show()
    D_noisy = nrm_clp.apply_noise(D)
    plt.title("NormalResamplingNoise Clipped")
    plt.imshow(D_noisy)
    plt.show()
    
def test_synth_D():
    for var in [1, 10, 20, 50, 100, 200, 400, 1000, 3000]:
        D = st.SynthELOTournamentSource(16, n_games=50, comp_var=var).init_D()
        plt.title("Synthetic ELO Tournament\nVar(competitor ELO)=%d"%var)
        plt.imshow(D)
        plt.show()
```

```python
test_synth_D()
```
